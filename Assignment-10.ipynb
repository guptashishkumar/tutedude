### Pandas Assignment

```python
import pandas as pd

# ----------------   TASK-1 --------------------
print("----------------   TASK-1 --------------------")
# Create a pandas series from list
marks=[78,85,90,66,72]
marks=pd.Series(marks, dtype='int32')
print("Series:")
print(marks)
print("First element: ",marks[0])
print("Last two element: ",marks[3:].values)
```

```python
# ----------------   TASK-2 --------------------
print("----------------   TASK-2 --------------------")
# add 5 garce mark to all studnets

marks= marks+5

print("Marks after grace of 5: ", marks.values)

marks=marks-2
print("Marks after -2: ", marks.values)

marks=marks*1.05
print("Marks after *1.05: ", marks.values)

marks=marks/2
print("Marks after /2: ", marks.values)
```

```python
# ----------------   TASK-3 --------------------
print("----------------   TASK-3 --------------------")

# Again initialising the mark
marks=[78,85,90,66,72]
marks=pd.Series(marks, dtype='int32')
print("Series:", marks.values, type(marks))

print("Max in Marks: ", marks.max())
print("Min in Marks: ", marks.min())
print("Sum in Marks: ", marks.sum())
print("Mean in Marks: ", marks.mean())

print("Passed Student", marks[marks>=70].values)

print("Count of Passed Student:", marks[marks>=70].value_counts().size)
```
